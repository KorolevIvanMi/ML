```python
import numpy as np
import matplotlib.pyplot as plt
```

далее я сгенерирую пару функций косинусойд и выведу их на один график. Хранить все данные буду в двумерном массиве
1. cos(x)
2. cos(x^2 + 2x - 5)
3. cos(1/2x *x^3)

```python
array_x = np.arange(-5,5,0.01)
array_y = np.zeros((3, array_x.shape[0]))
```

```python
array_y[0] = np.cos( array_x)
array_y[1] = np.cos(array_x**2 + 2*array_x -5) -2
array_y[2] = np.cos((1/array_x) * array_x**3) + 2
```

```python
plt.hexbin(array_x, array_y[0], cmap= "plasma", gridsize=100, alpha = 1)
plt.hexbin(array_x, array_y[1],  cmap= "plasma", gridsize=100, alpha = 1)
plt.hexbin(array_x, array_y[2],  cmap= "plasma", gridsize=100, alpha = 1)
plt.title("Различные косинусойды")
# plt.grid(True)
plt.show()
```
