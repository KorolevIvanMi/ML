```python
array_for_plot_x = np.arange(-10, 10, 0.1)


array_for_plot_y = np.sin(array_for_plot_x)

plt.plot(array_for_plot_x, array_for_plot_y)
plt.show()
```

```python
points = np.arange(-5, 5.2, 0.2) 
xs, ys = np.meshgrid(points, points)
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(xs, ys, s=3) # s=3 для маленьких точек
plt.title('Meshgrid of Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
```

```python
z = np.sqrt(xs**2 + ys**2)
plt.imshow(z, cmap=plt.cm.viridis, extent=[-5, 5, -5, 5])
plt.colorbar()
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```
